```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv("titanic.csv")
display(df)
```

```python
survival_rate = df.groupby('Pclass')['Survived'].mean()
plt.figure(figsize=(8,5))
plt.bar(survival_rate.index, survival_rate.values)
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class (Pclass)")
plt.ylabel("Survival Rate")
plt.xticks([1, 2, 3])
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
survival_counts = df['Survived'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
    survival_counts,
    labels=["Non-Survived", "Survived"],
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Proportion of Survivors vs. Non-Survivors")
plt.show()
```

```python
grouped = df.groupby(["Pclass", "Sex", "Survived"]).size().unstack(fill_value=0)

grouped.plot(kind="bar", stacked=True, color=["lightcoral", "lightgreen"])
plt.xlabel("Passenger Class & Sex")
plt.ylabel("Number of Passengers")
plt.title("Survivors vs Non-Survivors by Pclass & Sex")
plt.legend(["Non-Survived", "Survived"])
plt.show()
```

```python
survival_rate = df.groupby("Pclass")["Survived"].mean()

plt.bar(survival_rate.index, survival_rate.values, color="orange")
plt.xlabel("Passenger Class (Pclass)")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Passenger Class")
plt.show()
```
